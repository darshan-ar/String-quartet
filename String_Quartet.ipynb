```python
import torch
from scipy.io import wavfile
import torch.utils.data as Data
import torch.nn as nn
import torch.optim as optim
import numpy as np
```

```python
types = ['_delay_1','_delay_2','_delay_long','_noise_1','_noise_2']
instruments = ['cello','viola','violin_1','violin_2']
samples = 100

labels = []

y = []

for instrument in instruments:
    for i in types:
        filename = instrument+i
        _,waveform = wavfile.read('data\\'+instrument+'\\'+filename+'.wav')
        waveform = waveform.astype('float32')
        waveform /= 255
        waveform =np.expand_dims(torch.tensor(waveform),axis=0)
        print(np.shape(list(waveform)))
        y.append(list(waveform))
        filename_original = "data\\"+instrument+"\\"+instrument+"_min.wav"
        _,waveform_original = wavfile.read(filename_original)
        waveform_original = waveform_original.astype('float32')
        waveform_original/=255
        waveform_original=np.expand_dims(torch.tensor(waveform_original),axis=0)
        labels.append(list(waveform_original))
print(np.shape(labels),np.shape(y))
```

```python
labels = np.repeat(labels, samples,axis=0)
y = np.repeat(y, samples,axis=0)

print(np.shape(y),np.shape(labels))
```

```python
BATCH_SIZE = 20
torch.manual_seed(1)
epochs=100


y = torch.from_numpy(y)
labels = torch.from_numpy(labels)
dataset_ai = Data.TensorDataset(y,labels)

loader = Data.DataLoader(

    dataset= dataset_ai,

    batch_size=BATCH_SIZE,

    shuffle=True,

    num_workers=0,

)



class DiscreteNet(torch.nn.Module):
        # use an embedding layer to build a NN

    def __init__(self):

        super(DiscreteNet, self).__init__()

        self.embedding = nn.LSTM(2646000,1)
        self.embedding2 = nn.Linear(1,2646000)


    def forward(self, x):

        y,x = self.embedding(x)
        y = self.embedding2(y)
        return y



f = DiscreteNet()


optimizer = optim.Adam(params=f.parameters(), lr=0.01)
loss = nn.MSELoss()


def run():
    for epoch in range(epochs):

        for batch_x,l_1 in loader:
            optimizer.zero_grad()
            print(np.shape(batch_x))
            fx = f(batch_x)
            l = loss(fx,l_1)
            l.backward()
            optimizer.step()
            print("Epoch: ",epoch,"Loss: ",l.item())

    print('Finished Training')


run()
```
